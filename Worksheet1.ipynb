• For the following arrays:
A = np.array([[3,4],[7,8]]) and B = np.array([[5,3],[2,1]]);
Prove following with Numpy:
1. Prove A.A−1 = I.

```python
import numpy as np

A = np.array([[3, 4], [7, 8]])

A_inv = np.linalg.inv(A)
result = np.dot(A, A_inv)

print(result)
```

2. Prove AB ̸= BA.

```python
import numpy as np

A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])

AB = np.dot(A, B)
BA = np.dot(B, A)

print("A*B =\n", AB)
print("B*A =\n", BA)
print("Are A*B and B*A equal?", np.array_equal(AB, BA))
```

3. Prove (AB)

T = BTAT

```python
import numpy as np

A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])

AB = np.dot(A, B)
AB_T = AB.T

B_T = B.T
A_T = A.T
BTA_T = np.dot(B_T, A_T)

print("Transpose of AB:\n", AB_T)
print("B.T * A.T:\n", BTA_T)
print("Are they equal?", np.array_equal(AB_T, BTA_T))
```

• Solve the following system of Linear equation using Inverse Methods.

2x − 3y + z = −1
x − y + 2z = −3
3x + y − z = 9

{Hint: First use Numpy array to represent the equation in Matrix form. Then Solve for: AX = B}

```python
import numpy as np

A = np.array([[2, -3, 1],
              [1, -1, 2],
              [3, 1, -1]])

B = np.array([-1, -3, 9])

X = np.dot(np.linalg.inv(A), B)

print(X)
```

• Now: solve the above equation using np.linalg.inv function.{Explore more about ”linalg” function
of Numpy}

```python
import numpy as np

A = np.array([[2, -3, 1],
              [1, -1, 2],
              [3, 1, -1]])

B = np.array([-1, -3, 9])

A_inv = np.linalg.inv(A)
X = A_inv.dot(B)

print(X)
```
